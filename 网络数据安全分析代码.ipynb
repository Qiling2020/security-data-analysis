# 文件输入-黑名单资料
实验室的数据包含在主机探测.csv并具有列：主机、检测和检测计数。

```python
import pandas as pd

df = pd.read_csv('E:/ql/host_detections.csv', names=['host','detections','detection_count'], header=1, encoding='utf-8')
```

# 清除-黑名单数据
在df数据帧上删除列主机的重复项

```python
df.drop_duplicates(subset='host', inplace=True)
```

```python
black_list_sources = set()
def get_list(x):
    detections = []
    if not (len(x) == 1 and int(x) == 0):
        x = x.replace(" ", "")
        x = x.replace("u'", "")
        x = x.replace("'", "")
        x = x.replace("[", "")
        x = x.replace("]", "")
        [black_list_sources.add(i) for i in x.split(',') if len(i) > 1]
        [detections.append(i) for i in x.split(',') if len(i) > 1]
    return detections
df.detections = df.detections.apply(lambda x: get_list(x))
```

将得到的多维列表连接到现有数据帧的“边”。
您可以看到主机02b123c.netsolhost.com有0个检测，所有黑名单值都有0个。其中0lilioo0l0o00lill.info有7个检测，1个代替它的检测（例如hpHosts）。
